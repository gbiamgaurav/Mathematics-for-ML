### L**2 Norm

```python
x = np.array([12,5,6])
x
```

```python
(12**2+5**2+6**2)**(1/2)
```

```python
np.linalg.norm(x)
```

```python
xpt = torch.tensor([12,5,6.])
xpt
```

```python
torch.norm(xpt)
```

```python
xtf = tf.Variable([12,5,6.])
xtf
```

```python
tf.norm(xtf)
```

### Distance Formula

```python
x = np.array([12,5,6])
x
```

```python
y = np.array([3,4,9])
y
```

```python
((12-3)**2+(5-4)**2+(6-9)**2)**(1/2)
```

```python
np.linalg.norm(x-y)
```

```python
xpt = torch.tensor([12,5,6.])
ypt = torch.tensor([3,4,9.])
```

```python
torch.norm(xpt-ypt)
```

```python
xtf = tf.Variable([12,5,6.])
ytf = tf.Variable([3,4,9.])
```

```python
tf.norm(xtf-ytf)
```
