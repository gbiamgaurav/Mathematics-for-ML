### Dot Product

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
x = np.array([2,3,5])
```

```python
x
```

```python
xpt = torch.tensor([2,3,5.])
xtf = tf.Variable([2,3,5.])
```

```python
y = x+1
y
```

```python
np.dot(x,y)
```
